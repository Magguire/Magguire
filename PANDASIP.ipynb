## Data Analysis

### MTN C'orte Cote d'Ivoire Datasets Analysis

```python
#From the analysis below, Voice is the product that is mostly required
df1['Billing Price'].groupby(df1['Product']).sum()

# Voice seems to be the most essential service as it's bill is relatively high.
```

```python
# The average price per product is as fillows:

df1['Billing Price'].groupby(df1['Product']).mean()

# The average billing price for voice is 55.169961, data is 7.628861 and sms is 7.555791.
```

```python
# The standard deviation for price per product is as fillows:

df1['Billing Price'].groupby(df1['Product']).std()

# The data below tells us that the price of voice product is widely spread from the average billing price between Cities.
# There are Cities that have a high voice billing price, others fair and others have a low billing price.
# The billing price of data Service is relatively spread out.
# The sms product price range is minimal compared to voice and data products.
```

```python
# The variance for the price per product is as follows:
df1['Billing Price'].groupby(df1['Product']).var()

#The variance shows the average degree to which each of the voice, data, sms billings differ from the mean. Voice billing are very widespread compaared to the rest.
```

```python
#The skewness for the price per product is as follows:
df1['Billing Price'].groupby(df1['Product']).skew()

#The data below tell us that the billing price is more on the higher side. 
```

```python
# Group data to products billing prices per day
DayVsproduct = df1.groupby(['Date','Product'])['Billing Price'].sum()
DayVsproduct

# Voice billing price is the highest per day hence the most essential service, followed by sms then data.
```

### Cellphone Geography Analysis

```python
# The most used City
Cities = list(df['City'])
Cities_dict = dict()
for city in Cities:
  occurences = Cities.count(city)
  Cities_dict[city] = occurences
  Cities.remove(city)

df = pd.DataFrame() 
df[['City','Occurence']] = [Cities_dict.keys(),Cities_dict.values()]
df.sort_values(by='Occurence', ascending=0)

# The most used city is Cocody which appears in the dataset 179 times, followed by Yopougon then Abobo and so on and the so forth.
# The least used city is Dananon which hasonly been used once.
```
